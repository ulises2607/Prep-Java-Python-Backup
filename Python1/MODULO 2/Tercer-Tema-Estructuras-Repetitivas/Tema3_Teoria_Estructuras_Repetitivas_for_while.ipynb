###**Estructuras Repetitivas**

Permiten repetir una serie de acciones, según se cumpla una condición, cuando la condición no se cumpla la repetición termina.
Cada repetición se llama bucle o ciclo y en programación, es una secuencia de instrucciones de código que se ejecuta repetidas veces, hasta que la condición asignada a dicho bucle deja de cumplirse. También los bucles los llamamos iteraciones.

###**Generar Números con Función Range**

**range**: representa una secuencia de números enteros. Se utiliza pasándole un úncio valor de *stop*. genera una secuencia de números enteros consecutivos entre 0 y stop-1.

Si se utiliza pasándole dos argumentos, *start* y *stop*, genera una secuencia de números enteros consecutivos entre start y stop-1.

La inclusión de un tercer argumento *step* (salto) fuerza a que la secuencia de números se genere con un salto de "step" entre un número y el siguiente.

````python
#range solo 
list(range(10))
````

````python
#genera numero desde 1 hasta 9
list(range(1,10))
````

````python
#generar numeros 1, 3, 5, 7, 9 numeros impares entre 1 y 10
list(range(1,10,2))
````

````python
#generar numeros 0, 2, 4, 6, 8 numeros impares entre 0 y 10
list(range(0,10,2))
````

###**Estructura for**
*La sentencia **for** repiten una porción de código para un conjunto de valores.*

La sentencia for de Python itera sobre los ítems de un conjunto de datos (cd)



```
for <nombre> in <cd>:
  <bloqute de código>
```
**nombre** es una variable, que creamos nosotros,
que va tomando todos los valores de **cd**, uno por uno, en cada iteración. La idea es usar el valor que toma la variable dentro del bloque de código. La cantidad de iteraciones viene dada por la cantidad de elementos del conjunto de datos.




````python

#range generara desde 0
# hasta 3, si pasamos un 4
for i in range(4):
    print(i)
````

````python
#range genera numeros de 1 a 4, 
#el primer valor es el valor inicial 
#y el ultimo valor es el final
for i in range(1,4):
    print(i)
````

````python
#mostrar un mensaje 3 veces
for i in range(1,4):
    print('hola python')
````

````python
#podesmos generar numeros
# desde 1 hasta 10 pero
# que i se incremente de a 2
for i in range(1,11,2):
    print(i)
````

````python
#mostrar el resultado de sumar 5 números impares consecutivos

c = 0
for num in range(1,10,2):
  c = c + num
print (f"la suma es {c}")
print(" ")
````

###**contadores y Acumuladores**

####**Contadores:** 

Es la acción de contar que tenemos en la vida real, por ejemplo cuando llegamos a una fila de cajero contar a la personas que hay adelante. Los contadores son siempre variables enteras. Se inicializan en cero. Cuando quiero incrementar o contar debemos actualizar la variable en 1. Aunque no siempre es 1 va haber veces en que la podemos incrementar en mas de 1 por ejemplo Cuando nos piden contar en algún momento de la app debemos mostrar el valor parcial o final de la cuenta. Ejemplo

````python
#1 - se crea una variable entera en cero
cont = 0
#2 - se utiliza cuando deseamos contar
cont = cont +1
#3 - se muestra al final o cuando se desee mostrar el valor del contador
print(cont)
````

````python
#un solo contador para numeros positivos
contPos=0
N = int(input('ingrese la cantidad de repeticiones que desea:'))
for i in range(1,N+1):
    num = int(input('ingrese un numero:'))
    if num>0:
       print(num,' es positivo')
       contPos=contPos+1
    else:
      if num<0:
          print(num,' es negativo')
      else:
          print(num, ' es cero')

print('la cantidad de positivos es:',contPos)
````

````python
#puedo crear mas de un contador para diferente cuentas
contPos=0
contNeg=0
contCero=0
N = int(input('ingrese la cantidad de repeticiones que desea:'))
for i in range(1,N+1):
    num = int(input('ingrese un numero:'))
    if num>0:
       print(num,' es positivo')
       contPos=contPos+1
    else:
      if num<0:
          print(num,' es negativo')
          contNeg=contNeg+1
      else:
          print(num, ' es cero')
          contCero=contCero+1

print('la cantidad de positivos es:',contPos)
print('la cantidad de negativos es:',contNeg)
print('la cantidad de ceros es:',contCero)
````

####**Acumuladores:**


Un acumulador es una variable, no necesariamente entera, pero sí numérica.

Su objetivo es “acumular”, es decir: acopiar, almacenar, añadir un cierto valor.

La diferencia con una variable cualquiera es que el acumulador agrega un nuevo valor al que ya tiene


Ejemplo


````python
#1 - se crea una variable en cero
suma = 0
valor = 10
#2 - se utiliza cuando deseamos almacenar algun valor en el acumulador
#valor seria la variable o el dato que queremos que se sume al acumulador
suma = suma + valor
#3 - se muestra al final o cuando se desee mostrar el valor del acumulador
print(suma)
````

````python
#puedo crear mas de un acumulador para diferente cuentas
sumaPos=0
N = int(input('ingrese la cantidad de repeticiones que desea:'))
for i in range(1,N+1):
    num = int(input('ingrese un numero:'))
    if num>0:
       print(num,' es positivo')
       sumaPos=sumaPos+num #se esta acumulando el num
    else:
      if num<0:
          print(num,' es negativo')
      else:
          print(num, ' es cero')

print('la suma de positivos es:',sumaPos)
````

````python
#puedo crear mas de un acumulador para diferente acumulaciones
sumaPar=0
sumaImpar=0
N = int(input('ingrese la cantidad de repeticiones que desea:'))
for i in range(1,N+1):
    num = int(input('ingrese un numero:'))
    if num%2==0:
       sumaPar=sumaPar+num
    else:
       sumaImpar=sumaImpar+num

print('suma de numeros pares:', sumaPar)
print('suma de numeros impares:', sumaImpar)
````

###**Estructura while**

La sentencia o ciclo *while* es una sentencia de control de flujo que se utiliza para ejecutar un bloque de código de forma repetitiva mientras se cumpla una condición determinada.

La estructura de esta sentencia es la siguiente:



```
while <condición>:
    <bloque de código>
```
De la misma forma que la sentencia if, si la condición evalúa a True se ejecutará el bloque de código. Sino, se ignorará. La diferencia está en que al terminar de ejecutar el bloque, automáticamente se volverá a comprobar la condición, si continúa siendo verdadera se ejecutara el bloque nuevamente y asi sucesivamente. Cuando la condición evalúe a False, se ignorará el bloque de código y se continuará la ejecución normal.

`Cada ejecución del bloque de código se denomina iteración.`



````python
#repetir accciones a pedido del operador
respuesta = 'si'
while respuesta =='si':
    print('hola python  - while')
    respuesta = input('desea continuar si o no ? ')
````

para controlar el while mediante una variable de contador debemos crear una variable entera en un valor inicial y que por cada giro se vaya incrementando, esta variable tambien debe ser usada en la condicion del while.

````python
#repetir ingresar N datos, la variable contador es i
#una variable entera que se crea y por cada giro se va incrementando
N = int(input('ingrese la cantidad de repeticiones que desea:'))
i=1
while i<=N:
    print('hola python  - while - variable de contador')
    i=i+1
````

````python
#repetir ingresar N datos, la variable contador es i una variable 
#entera que se crea y por cada giro se va decrementando
N = int(input('ingrese la cantidad de repeticiones que desea:'))
i=N
while i>=1:
    print('hola python  - while - variable de contador')
    i=i-1
````

para controlar el while mediante una variable de bandera o logica debemos crear una variable logica en un valor inicial, y usar la variable en la condicion, luego en algunos de los giros de la repeticion la varible tiene que cambiar de estado.

````python
#repetir hasta que se ingrese un numero cero -
#usando variable bandera o logica
band=True #variable de bandera
while band==True:
    num = int(input('ingrese un numero:'))
    if num==0:
       band=False    
````

Ejemplo usando contadores y acumuladores

````python
'''
determine si un número es positivo, negativo o cero, 
el proceso finaliza cuando el operador no desea continuar.
contar lo negativos => contador   +1
sumar los positivo => acumulador  + valor
'''
contNeg=0
sumaPos=0
respuesta = 'si'
while respuesta =='si':
    numero = int(input('Ingrese un numero:'))
    if numero>0:
        print('es positivo')
        sumaPos=sumaPos+numero
    else:
        #neg cero 
        if numero==0:
            print('es cero')   
        else:
            #negativo
            print('negativo')
            contNeg=contNeg+1
    respuesta = input('desea continuar si o no ? ')
print('la cantidad de negativo es:', contNeg)
print('la suma de positivos es:',sumaPos)
````
