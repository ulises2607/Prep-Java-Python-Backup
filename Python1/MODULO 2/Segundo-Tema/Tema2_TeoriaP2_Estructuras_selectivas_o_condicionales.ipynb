###**Estructuras selectivas o condicionales**

En un programa siempre hay que tomar decisiones por lo usaremos una estructura que nos permita armar caminos de acuerdo a una condicin.
Siempre en un cambio hay que realizar acciones ya sea una o las que deseemos.


####**Indentacin en la estructura de control**

Python utiliza la indentacin para delimitar la estructura permitiendo establecer bloques de cdigo. No existen comandos para finalizar las lneas ni llaves con las que delimitar el cdigo. Los nicos delimitadores existentes son los dos puntos ( **:** ) y la indentacin del cdigo.

La indentacin es un espacio en blanco o sangrado con lo que se indicara el inicio del bloque, si en las posteriores lneas no introdujeramos el sangrado, significa el final de dicho bloque de cdigo, con lo cual para finalizar un bloque de cdigo, slo tenemos que dejar de introducir el sangrado, no tenemos que usar ninguna llave ni smbolo. 

Esta regla se aplica para las estructuras de control **IF - IF ELSE - IL ELIF** a la hora de implementarlas

```python
# Ejemplos
a = 10
b = 5
if a > b: #con los ":" idica que empieza un bloque condicional en este caso
  print(str(a) + "es mayor que "+ str(b) ) #esto es un bloque
# iniio del bloque dejando una indentacin
```

### **Estructura selectiva simple if**

Consta de crear una condicin si la condicin es true se entra en el camino y se ejecutan las acciones. En caso que sea false la condicin no se ejecuta nada.

Screenshot_8.jpg


Ejemplo


```python
if condicion:
    acciones
```

```python
edad = int(input('Ingrese su edad:'))
if edad>=18:
    print('es mayor de edad')
```

```python
numero = int(input('Ingrese su numero:'))
if numero>0:
    print('es positivo')
```

```python
numero = int(input('Ingrese su numero:'))
if numero>=1 and numero<=10:
    print('esta entre 1 y 10')
```

### **Estructura selectiva if else**

Es una estructura selectiva doble donde si la condicin es True se entre por el camino y se ejecutan sus acciones. Pero si la condicin es False se crea otro camino contrario a la condicin donde podremos ejecutar acciones diferentes.
image.png

Ejemplo


```python
if condicion:
   acciones_camino_if
else:
   accciones_camino_else
```

```python
edad = int(input('Ingrese su edad:'))
if edad>=18:
    print('es mayor de edad')
else:
    print('no es mayor de edad')
```

```python
numero = int(input('Ingrese su numero:'))
if numero>=1 and numero<=10:
    print('esta entre 1 y 10')
else:
    print('no esta entre 1 y 10')
```

```python
numero = int(input('Ingrese su numero:'))
if numero%2==0:
    print('es par')
else:
    print('no es par o es impar')
```

```python
apellido=input('Ingrese apellido:')
if apellido == 'mamani':
    print('es mamani')
else:
    print('es un apellido que no es mamani')
```

###**Estructura if elif**

Se trata de una estructura selectiva mltiple o una estructura de if anidados donde podremos armar ms de dos camino. Y dentro de un if o else se pueden poner otro if.
image.png
Ejemplos:


```python
if condicion1:
  #acciones condicion1
elif condicion2:
  #acciones condicion2
elif condicion3
  #acciones condicion3
else:
  #acciones else
```

```python
 #en este ejemplo dentro de un else hay otro if
numero = int(input('Ingrese un numero:'))
if numero>0:
    print('es positivo')
else:
    #neg cero 
    if numero==0:
        #es cero
        print('es cero')   
    else:
        #negativo
        print('negativo')
```

```python
#en este ejemplo el dentro de un if hay otro if
numero = int(input('Ingrese un numero:'))
if numero>=1 and numero<=3 :
    print('esta entre 1y 3')
    if numero==1:
      print('uno')
    else:
        if numero==2:
            print('dos')
        else:  
            if numero==3:
                print('tres')

else:
    print('No es 1, 2 ni 3')
```

```python
#este es una selectiva multiple
opcion=input('Ingrese una opcion:')
if opcion=='a':
    print('aqui va la suma')
elif opcion=='b':
      print('aqui va la resta')
elif opcion=='c':
      print('aqui va la factorial')
elif opcion=='d':
      print('fin de la app')
else:
      print('opcion no valida')
```
