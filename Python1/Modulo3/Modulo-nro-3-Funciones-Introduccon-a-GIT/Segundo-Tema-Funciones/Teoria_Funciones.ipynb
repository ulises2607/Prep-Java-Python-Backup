```python
def sumar(a,b):			#Definimos la funcin sumar
  x = a + b
  return x
def restar(a,b):			#Definimos la funcin restar
  x = a - b
  return x
def multiplicar(a,b):		#Definimos la funcin multiplicar
  x = a * b
  return x
def dividir(a,b):			#Definimos la funcin dividir
  x = a / b
  return x
# Seccin 2 ################################################
operar = True
while operar: 			#Creamos un ciclo
  a = int(input("Ingresa el primer numero: \n")) 
  b = int(input("Ingresa el segundo numero: \n"))   
  op = input ("""Elegir clculo
       1- Sumar
       2- Restar
       3- Multiplicar
       4- Dividir \n""")
# Seccin 3 ################################################
  if op == '1':		#Llamamos a funcin sumar
    res = sumar(a,b)
    print("El resultado es: ",res)
  elif op == '2':		#Llamamos a funcin restar
    res = restar(a,b)
    print("El resultado es: ",res)
  elif op == '3':		#Llamamos a funcin multiplicar
    res = multiplicar(a,b)
    print("El resultado es: ",res)
  elif op == '4':		#Llamamos a funcin dividir
    res = dividir(a,b)
    print("El resultado es: ",res)
  else:
    print("""Nmero de opcin errneo""")
# Seccin 4 #################################################
  s = input("Quieres hacer otra operacin? S/N \n")
  if (s.upper() == 'N'):
    operar = False		# Para finalizar el ciclo
print("FIN de calculadora") 	# Mensaje luego que corta el ciclo
```

## PARMETROS POR VALOR Y PARMETROS POR REFERENCIA

Una funcin puede definir, opcionalmente, una secuencia de parmetros que poseen una caracterstica muy importante.

	**Paso por valor**: Un lenguaje de programacin que lo utiliza, lo que hace es copiar el valor de las variables en los respectivos parmetros. Cualquier modificacin del valor del parmetro, no afecta a la variable externa correspondiente.

	**Paso por referencia**: Un lenguaje de programacin que lo utiliza, lo que hace es copiar en los parmetros la direccin de memoria de las variables que se usan como argumento. Hacen referencia al mismo objeto y cualquier modificacin del valor en el parmetro afectar a la variable externa correspondiente.

**En Python, no aplica este consepto de forma literal, Ya que no realiza copias, sino crea NUEVAS VARIABLES o hacer referencia a la varible si esta es mutable** 

Veamos un ejemplo de paso por valor.


```python
def suma(a,b):
  a = a + 1
  b = b + 1
  print(f"Valores de a y b dentro de la funcin: {a},{b}")
  print(f"Identidicadores de las variables a y b dentro de la funcin: {id(a)}, {id(b)}")
  return a+b

a = 5
b = 8
print(f"Identidicadores de las variables a y b fuera de la funcin: {id(a)}, {id(b)}")
print(f"resultado de la funcin: {suma(a,b)}")

#Van a notar que los identificadores son distintos,
#no hace una copia del valor sino que crea un identificador nuevo apuntando a ese valor
```

```python
def suma(lista2):
  lista2[0] += 1
  lista2[1] += 1
  print(f"Identidicadores de la variable lista2 dentro de la funcin: {id(lista2)}")
  return lista2[0]+lista2[1]

lista = [5,8]
print(f"Identidicadores de la variable lista fuera de la funcin: {id(lista)}")
print(f"resultado de la funcin: {suma(lista)}")

#En este caso lista2 tiene el mismo identificador que lista 
```

##**mbito de una variable**

**El mbito de una variable es el contexto en el que existe esa variable**. As, una variable existe en dicho mbito a partir del momento en que se crea y deja de existir cuando desaparece su mbito.

Adems, las variables son accesibles solo desde su propio mbito, pero con alguna excepcin, como veremos ms adelante.

Los principales tipos de mbitos en Python son dos:


*   **mbito local:** corresponde con el mbito de una funcin, que existe desde que se invoca a una funcin hasta que termina su ejecucin. En un programa, el mbito local corresponde con las lneas de cdigo de una funcin. Dicho mbito se crea cada vez que se invoca a la funcin. Cada funcin tiene su mbito local. No se puede acceder a las variables de una funcin desde fuera de esa funcin o desde otra funcin.

*   **mbito global:** corresponde con el mbito que existe desde el comienzo de la ejecucin de un programa. Todas las variables definidas fuera de cualquier funcin corresponden al mbito global, que es accesible desde cualquier punto del programa, incluidas las funciones. Si desde un mdulo A importamos un mdulo B mediante un import, desde A podremos acceder a las variables globales de B.


###**Variables locales**

En Python las variables locales son aquellas definidas dentro de una funcin. Solamente son accesibles desde la propia funcin y dejan de existir cuando esta termina su ejecucin. Los parmetros de una funcin tambin son considerados como variables locales.

Desde el cuerpo principal del programa no se puede acceder a las variables locales de ninguna funcin. Tampoco es posible acceder a las variables locales de una funcin desde otra.



```python
def saludar():
	saludo = 'Hola!' #esta variable local pertenece al mbito local de la funcin saludar
	print(saludo)


def sumar(sumando1, sumando2):
	#sumando1, sumando2 y suma son variables locales que corresponden al mbito local de la funcin sumar
	suma = sumando1 + sumando2

	return suma


saludar()
print(sumar(4, 5))
```

###**Variables globales**
Las variables globales en Python son aquellas definidas en el cuerpo principal del programa fuera de cualquier funcin. Son accesibles desde cualquier punto del programa, incluso desde dentro de funciones. Tambin se puede acceder a las variables globales de otros programas o mdulos importados.

Por esto decimos que las variables globales son aquellas de mbito global. Te enseo un ejemplo donde se puede observar cmo se permite el acceso a las variables globales desde dentro de una funcin:

```python
def saludar():
	print(saludo)


saludo = 'Hola, Mundo!' #variable global definida en el cuerpo principal del programa
saludar()
```

###Cmo modificar una variable global desde una funcin


```python
contador = 10

def reiniciar_contador():
	contador = 0


print(f'Contador antes es {contador}')
reiniciar_contador()
print(f'Contador despus es {contador}')
```

En el ejemplo anterior tenemos una variable global contador que en un caso real nos servira para contar cualquier cosa. Tambin disponemos de una funcin que nos permite reiniciar el contador cuando queramos. Al ejecutar el cdigo lo esperable sera obtener por pantalla los valores 10 y 0, pero en su lugar tenemos los valores 10 y 10.

Esto sucede as no por capricho, sino porque es un mecanismo de proteccin para evitar modificar sin querer una variable global que posiblemente terminara alterando el funcionamiento normal de un programa.

Para ganar acceso de modificacin es necesario utilizar el modificador global. Con esta declaracin le estamos diciendo a Python que sabemos que vamos a utilizar una variable global y que queremos modificarla. Veamos el ejemplo corregido y su ejecucin:

```python
contador = 10

def reiniciar_contador():
	global contador
	contador = 0


print(f'Contador antes es {contador}')
reiniciar_contador()
print(f'Contador despus es {contador}')
```

##  **ARGS vs KWARGS**

Python nos permite crear funciones que acepten un nmero indefinido de parmetros sin necesidad de que todos ellos aparezcan en la cabecera de la funcin. Los operadores * y ** son los que se utilizan para esta funcionalidad.

## **Args**

El parmetro especial ***args** en una funcin se usa para pasar, de forma opcional, un nmero variable de argumentos posicionales.
* Quien indica que el parmetro es de este tipo es el smbolo *, el nombre* **args** se usa por convencin.
* El parmetro recibe los argumentos como una tupla.
* Es un parmetro opcional.
* El nmero de argumentos al invocar a la funcin es variable.
* Son parmetros posicionales, es decir, su valor depende de la posicin en la que se pasen a la funcin.

La siguiente funcin toma dos parmetros y devuelve la suma de los mismos:


```python
def sum(x, y):
   return x + y

###Si llamamos a la funcin con los valores 2 y 3, el resultado devuelto ser 5.
```

Pero, si necesitamos sumar un valor ms?

```python
sum(2, 3, 4)
```

Por supuesto que esperbamos que la llamada falle. Una posible solucin sera agregar parmetros, pero tampoco sabramos cuntos.

La mejor solucin y ms elegante es hacer uso de ***args** en la definicin de esta funcin. De este modo, podemos pasar tantos argumentos como queramos.
 
Primero hay que redisear la funcin sum y con esa nueva implementacin, podemos llamar a la funcin con cualquier nmero variable de valores:


```python
def sum(*args):
   value = 0
   for n in args:
       value += n
   return value
 
#Invocaciones de la funcin
print(sum())
print(sum(4,5))
print(sum(4,5,24,11))
```

El comportamiento de la funcin siempre es el mismo, con independencia del nmero de argumentos que pasamos.

Con esto resolvemos nuestro problema inicial, en el que necesitbamos un nmero variable de argumentos.


## **Kwargs**
<b>kwargs</b> permite pasar argumentos de longitud variable asociados con un nombre o key a una funcin. Deberas usar <b>kwargs</b> si quieres manejar argumentos con nombre como entrada a una funcin.

Las principales diferencias con respecto ***args** son:

* Lo que indica que el parmetro es de este tipo es el smbolo **, el nombre <b>kwargs</b> se usa por convencin.
* El parmetro recibe los argumentos como un diccionario.
* Al tratarse de un diccionario, el orden de los parmetros no importa. Los parmetros se asocian en funcin de las claves del diccionario.

Veamos un ejemplo sencillo de cmo ** se asocia a un diccionario


```python
def fun(**params):
  print(params)
  
fun(x=5, y=8)
fun(x=5, y=8, z=4)
```

Es decir, dentro de la funcin no solo tenemos acceso a la variable como con * **args**, sino que tambin tenemos acceso a un nombre o key asociado

Tambin es posible que la cabecera de una funcin utilice uno o varios argumentos posicionales, seguidos del operador * o **. Esto nos proporciona bastante flexibilidad a la hora de invocar a una funcin. 

En el siguiente ejemplo tenemos 2 parmetros fijos y uno que permite agregar tantos como necesitemos.


```python
def print_record(nombre, apellido, **rec):
  print("Nombre: ", nombre)
  print("Apellidos:", apellido)
  for k in rec:
    print("{0}: {1}".format(k, rec[k]))
 
print_record("Juan", "Coll", edad=43, localidad="Madrid")
```

En resumen:

* Utiliza ***args** para pasar de forma opcional a una funcin un nmero variable de argumentos posicionales.

* El parmetro ***args** recibe los argumentos como una tupla.

* Emplea ****kwargs** para pasar de forma opcional a una funcin un nmero variable de argumentos con nombre.

* El parmetro ****kwargs** recibe los argumentos como un diccionario.

Por ltimo, utilizar ***args** y ****kwargs** ahorra tiempo y esfuerzo en cuanto a anlisis e implementacin dado que flexibiliza el diseo de funciones, pero su uso indiscriminado puede originar resultados inesperados cuya revisin y arreglo nos podra llevar ms tiempo an que el diseo estratgico original.

## **Librera datetime**
Esta libreria permite crear objetos para manejar fechas y horas por ejemplo:

```python
from datetime import datetime

dt = datetime.now()    # Fecha y hora actual

print(dt)
print(dt.year)         # ao
print(dt.month)        # mes
print(dt.day)          # da

print(dt.hour)         # hora
print(dt.minute)       # minutos
print(dt.second)       # segundos
print(dt.microsecond)  # microsegundos

print("{}:{}:{}".format(dt.hour, dt.minute, dt.second))
print("{}/{}/{}".format(dt.day, dt.month, dt.year))
```

Es posible crear un datetime manualmente pasando los parmetros (year, month, day, hour=0, minute=0, second=0, microsecond=0, tzinfo=None). Slo son obligatorios el ao, el mes y el da.

```python
from datetime import datetime

dt = datetime(2000,1,1)
print(dt)
```
